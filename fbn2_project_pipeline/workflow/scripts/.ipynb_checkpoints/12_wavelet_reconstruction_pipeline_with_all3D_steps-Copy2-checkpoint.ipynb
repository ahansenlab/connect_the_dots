## Run

```python
skip_existing = True
do_first_last = True #False
show_this_frame = [0,10]#[0,350]
do_dry_run = False#False
verbose = False
num_files_to_run = 0
do_plotting = True
for idx in [0]:
    try:
        # Load files
        input_file = os.path.join(in_pathnames[idx],in_filenames[idx])
        output_file = os.path.join(output_filepath,'Dot_predictions_'+in_filenames[idx][:-3]+'csv')      


        # Skip existing files
        if skip_existing == True:
            if os.path.exists(output_file) == True:
                print(f'Skipping {in_filenames[idx]}')
                continue

        # Dry run toggle
        if do_dry_run:
            assert(os.path.exists(input_file) == True)
            num_files_to_run += 1
            continue

        # Get CZI metadata 
        img_info, metadata = get_CZI_metadata(input_file,filepath=None,verbose=False)
        sizeX, sizeY, sizeZ, sizeT, num_channels = img_info
        print(f"Doing {idx}/{len(in_filenames)}: {in_filenames[idx]} \n {sizeT} frames")

        # Subset of images toggle
        if do_first_last == True:
            frames = show_this_frame#[0,sizeT//2,sizeT-1]
        else:
            frames = range(sizeT)


        # initialize stuff
        centroids3D_ch0 = []
        centroids3D_ch1 = []                    
        # dot_ID = []   
        frame_ID = []
        channel_ID = []
        X = []
        Y = []
        Z = []
        mean_intensity_list = []
        max_intensity_list = []
        mean_surrounding_list = []
        sum_intensity_list = []
        dot_size_pixels_list = []
        channels = [0,1]

        ## Iterate through each frame, each channel and retrieve dots
        for frame, ch in product(frames,channels):
            if verbose == True:
                print("Doing frame {}/{}, channel {}".format(frame,sizeT,ch))

            ## Retrieve Z-stack
            zstack = get_CZI_zstack(input_file,frame=frame,channel=ch,filepath=None,img_info=img_info) 

            ## Do Wavelet-decomposition of the image
            wavelet_type = channel_dict[ch]['wavelet_type']
            levels = channel_dict[ch]['wavelet_levels_to_remove']
            coeffs = pywt.wavedecn(zstack,wavelet_type)

            ## Reconstruct image, after filtering different wavelet coefficients 
            for level in levels:
                if level == 0:
                    coeffs[0] = np.zeros_like(coeffs[0])
                else:
                    coeffs[level] = {k: np.zeros_like(v) for k, v in coeffs[level].items()} #level 2 remove
            rec = pywt.waverecn(coeffs, wavelet_type) 

            ## reshape rec to match
            if (rec.shape != zstack.shape):
                if frame == 0:
                    print("Warning! Pixel sizes of z-stack are ODD! Cropping Wavelet reconstruction")
                rec = rec[0:sizeZ,0:sizeY,0:sizeX]


            ## Apply smoothing to the reconstructed z-stack
    #         dog_sigma1 = channel_dict[ch]['dog_sigma1']
    #         dog_sigma2 = channel_dict[ch]['dog_sigma2']
            image3d = rec#filter_zstack_DoG(rec, dog_sigma1 ,dog_sigma2, absolute_value=True) 


            ## Apply threshold to binarize image 
            stack_threshold = channel_dict[ch]['zstack_intensity_threshold']
            blobs = image3d.copy()
            blobs[blobs < np.percentile(blobs,stack_threshold)] = 0
            blobs[blobs>0] = 1

            ## Measure blobs/region properties 
            ### centroids of objects. i.e. (x,y,z) coordinates, volumes and intensities
            ### note that the values are actually returned as (z,y,x) coordinates
            blobs_labels = measure.label(blobs, background=0) 
            blob_metrics = measure.regionprops(blobs_labels, zstack) ## note that we use the unfiltered zstack as input
            centroids = [tuple(np.array(x.weighted_centroid,dtype=float)) for x in blob_metrics]
            vol = [x.area for x in blob_metrics]
            max_intensity = [x.max_intensity for x in blob_metrics]
            mean_intensity = [x.mean_intensity for x in blob_metrics]

            ## Filter localized dots based on size
            min_vol = channel_dict[ch]['min_dot_size_3D']
            max_vol = channel_dict[ch]['max_dot_size_3D']
            good_dots = [vi for vi, v in enumerate(vol) if ((v >= min_vol) and (v <= max_vol))]

            ## Update the list of localized dots
            X.extend([centroids[i][2] for i in  good_dots])
            Y.extend([centroids[i][1] for i in  good_dots])
            Z.extend([centroids[i][0] for i in  good_dots])
            frame_ID.extend([frame]*len(good_dots))
            channel_ID.extend([ch]*len(good_dots))
            mean_intensity_list.extend([mean_intensity[i] for i in  good_dots]) 
            max_intensity_list.extend([max_intensity[i] for i in  good_dots]) 
            dot_size_pixels_list.extend([vol[i] for i in  good_dots])


            ## Now plot things, if desired
            if (do_plotting == True) and (frame in show_this_frame):
                width_inches_x = 4; width_inches_y = 4; xwidth = 5; ywidth = 1
                fig, gs = _gridspec_inches(wcols = np.array([width_inches_x]*xwidth),\
                          hrows =np.array([width_inches_y]*ywidth),\
                         hspace=0.25,wspace=0.25)    

                count = 0
                plt.subplot(gs[count]); count +=1;
                plt.imshow(np.max(zstack,0),'gray')
                plt.title(f"Raw image - frame {frame} - channel {ch}")

                plt.subplot(gs[count]); count +=1;
                plt.imshow(np.max(rec,0),'gray')
                plt.title(f"Reconstructed image - frame {frame} - channel {ch}")     

                plt.subplot(gs[count]); count +=1;
                plt.imshow(np.max(image3d,0))
                plt.title(f"Filtered image - frame {frame} - channel {ch}")  

                plt.subplot(gs[count]); count +=1;
                plt.imshow(np.max(blobs,0))
                plt.title(f"Binarized image - frame {frame} - channel {ch}")  

                plt.subplot(gs[count]); count +=1;
                blobs_filt = np.zeros_like(blobs)#.copy()
                for v in good_dots:
                    blobs_filt[blobs_labels==(v+1)] = 1
                plt.imshow(np.max(blobs_filt,0))
                plt.title(f"Binarized + filtered image - frame {frame} - channel {ch}")  
                plt.show()

        # create Data Frame with the dots data
        df_particles = pd.DataFrame()
        df_particles['frame'] = frame_ID
        df_particles['channel'] = channel_ID
        df_particles['x'] = X
        df_particles['y'] = Y  
        df_particles['z'] = Z
        df_particles['mean_intensity_dot'] = mean_intensity_list
        df_particles['max_intensity_dot'] = max_intensity_list
        df_particles['dot_size_in_pixels'] = dot_size_pixels_list
    #     df_particles['sum_intensity_dot'] = sum_intensity_list

        if verbose == True:
            print("Saving output {}".format(output_file))

        if not do_first_last:
            df_particles.to_csv(output_file,index=False)
        
    except:
        print(f"failed at {output_file}")
```
