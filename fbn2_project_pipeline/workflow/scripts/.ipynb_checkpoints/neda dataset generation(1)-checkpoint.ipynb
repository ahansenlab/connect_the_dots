```python
def twoLocusMSD(t, Gamma, tau_c):
    """
    t : time coordinate
    Gamma : polymer prefactor
    tau_c : crossover-time to constant
    
    Note: we write 2 as prefactor, since  should be the anomalous
          diffusion coefficient of a single point on the polymer.
    """
    with np.errstate(divide='ignore'):
        ret = 2 * Gamma * (np.sqrt(t)*(1-np.exp(-tau_c/(np.pi*t))) + np.sqrt(tau_c)*scipy.special.erfc(np.sqrt(tau_c/(np.pi*t))))
    ret[np.isnan(ret)] = 0
    return ret

pixelsize = {'x' : 85, 'y' : 85, 'z' : 250}
# pixelsize = {'x' : 1, 'y' : 1, 'z' : 1}
def traj_noise_from_aggregate(noise2):
    voxel = np.array([pixelsize[dim] for dim in ['x', 'y', 'z']])
    noise_in_px = np.sqrt(noise2/np.sum(voxel**2))
    return noise_in_px*voxel
```

```python
# For localization error estimation we use only C36
msdfits = {}

data.makeSelection() # reset the tagset
for tag in data.tagset():
    data.makeSelection(tags=tag)
    
    reldata = data.process(lambda traj: traj.relative())
    print((tag,len(reldata)))
    
    msd, N = tl.analysis.MSD(reldata, giveN=True)
    # msd *= 1e-6
    t_msd = 20*np.arange(len(msd))

    popt, pcov = scipy.optimize.curve_fit(lambda x, G, tau, noise2 : np.log(twoLocusMSD(x, G, tau) + 2*noise2),
                                          t_msd[1:], np.log(msd[1:]), p0=(4e-3, 8e2, 0.04),
                                          sigma = 1/np.sqrt(N[1:]), bounds=(0, np.inf))
        
    msdfits[tag] = popt
    traj_noise = traj_noise_from_aggregate(popt[2])

    plt.loglog(t_msd[1:], msd[1:], label=tag)
    plt.loglog(t_msd[1:], twoLocusMSD(t_msd[1:], *popt[:2]) + 2*popt[2], color='r', label='fit')
    plt.loglog(t_msd[1:], twoLocusMSD(t_msd[1:], *popt[:2]), color='m', label='Rouse')
    plt.axhline(2*popt[2], color='c', label='error')

    plt.legend()
    plt.title('Fitting Rouse + localization error')
    plt.xlabel('time [seconds]')
    plt.ylabel('MSD [m^2]')
    plt.show()

    print("Fit results")
    print("-----------")
    print("   = {:.3g} m^2/s".format(popt[0]))
    print("   = {:.3g} s ==> rms(R_e2e) = {:.3g} m".format(popt[1], np.sqrt(popt[0]*np.sqrt(popt[1]))))
    print("  noise = {:.3g} m^2".format(popt[2]))
    print("    --> error on distance vectors: [{:.3f}, {:.3f}, {:.3f}]m".format(*traj_noise))
    print("    --> error on single spots:     [{:.3f}, {:.3f}, {:.3f}]m".format(*traj_noise/np.sqrt(2)))
    print("")
    print("relative uncertainty: {:.3g}".format(( np.sqrt(np.linalg.det(pcov)) / np.prod(popt) )**(1/3)))
    print("")
    
```

```python
model = neda.models.RouseModel(20, 1, 5, k_extra=1, looppositions=[(0, 0), (0, -1)])
noise = 0.1

def gen_Loopingtraces(N, T, t_loop, t_noloop, f_missing=0):
    assert type(N) == int
    assert type(T) == int
    assert 0 <= f_missing and f_missing <= 1
    lifetime = {0 : t_noloop, 1 : t_loop}
    
    for _ in range(N):
        states = np.zeros((T,))
        
        new_state = np.random.choice([0, 1])
        cur_t = np.random.exponential(lifetime[new_state])
        cur_ind = np.ceil(cur_t).astype(int)
        states[0:cur_ind] = new_state
        
        while cur_ind < len(states):
            new_state = 1-new_state
            cur_t = np.random.exponential(lifetime[new_state])
            new_ind = cur_ind + np.ceil(cur_t).astype(int)
            states[cur_ind:new_ind] = new_state
            cur_ind = new_ind
            
        if f_missing > 0:
            N_missing = np.random.binomial(len(states), f_missing)
            ind_missing = np.random.choice(np.arange(len(states)), size=N_missing, replace=False)
            states[ind_missing] = np.nan
            
        yield neda.Loopingtrace.fromStates(states, nStates=2)
        
def gen_Loopingtraces_1loop(N, T, t_loop, t_noloop, f_missing=0):
    assert type(N) == int
    assert type(T) == int
    assert 0 <= f_missing and f_missing <= 1
    
    for _ in range(N):
        states = np.zeros((T,))
        start = np.random.choice(T)
        states[start:min(T, start+t_loop)] = 1
            
        if f_missing > 0:
            N_missing = np.random.binomial(len(states), f_missing)
            ind_missing = np.random.choice(np.arange(len(states)), size=N_missing, replace=False)
            states[ind_missing] = np.nan
            
        yield neda.Loopingtrace.fromStates(states, nStates=2)
        
def parfun_trajectory_generation(loopingtrace):
    return model.trajectory_from_loopingtrace(loopingtrace, localization_error=noise, d=3)

def new_dataset(show_progress=True, gen_lt=gen_Loopingtraces, **kwargs):
    with Pool(20) as mypool:
        it = mypool.imap_unordered(parfun_trajectory_generation, gen_lt(**kwargs))
        if show_progress:
            it = tqdm(it, total=kwargs['N'])
            
        return tl.TaggedSet(it, hasTags=False)
```

# Utility: How much MCMC do we need?

```python
data = new_dataset(N=20, T=100, t_loop=20, t_noloop=20, f_missing=0.1)
MCMCconfigs = [
    {'iterations' :   300},
    {'iterations' :   500},
    {'iterations' :  1000},
    {'iterations' :  2000},
    {'iterations' :  3000},
    {'iterations' :  4000},
    {'iterations' :  5000},
    {'iterations' :  7000},
    {'iterations' : 10000},
]
model = neda.models.RouseModel(20, 1, 5, k_extra=1, looppositions=[(0, 0), (0, -1)])
priorfam = neda.ParametricFamily((0,), [(None, 0)])
priorfam.get = lambda logq : neda.priors.GeometricPrior(logq, nStates=2)

mcmc = neda.mcmc.TriMCMC(weights=[0.1, 1, 1], alpha=2)

it_ev_data = [] # (iterations, best evidences, data_inf)
def parfun(params):
    traj, conf = params
    conf['burn_in'] = 100
    return neda.main(traj, model, priorfam, conf, return_='traj')#, MCMCscheme=mcmc)
with Pool(20) as mypool:
    for conf in tqdm(MCMCconfigs):
        imap = mypool.imap_unordered(parfun, ((traj, conf) for traj in data))
        data_inf = tl.TaggedSet(imap, hasTags=False)
        best_evs = np.array([traj.meta['neda']['final']['evidence'] for traj in data_inf])
        it_ev_data.append((conf['iterations'], best_evs, data_inf))
```

```python
means = np.array([np.mean(ev) for _, ev, _ in it_ev_data])
stds  = np.array([np.std(ev)  for _, ev, _ in it_ev_data])
its = np.array([it for it, _, _ in it_ev_data])

plt.errorbar(its, means, stds)
plt.xlabel('#MCMC iterations')
plt.ylabel('final evidence, mean +- std')
plt.title('Convergence on 10 trajectories with T=100')
plt.show()
```
