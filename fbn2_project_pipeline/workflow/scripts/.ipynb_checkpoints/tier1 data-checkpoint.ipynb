## Loading data

```python
def load_clean_df(filename, pixelsize):
    df = pd.read_csv(filename)
    movie_files = np.unique(df.filename)
    df['fileid'] = np.where(df.filename.to_numpy()[:, np.newaxis] == movie_files[np.newaxis, :])[1]

    max_particle_id = np.max(df.particle)+1
    df['unique_particle_id'] = df.particle + max_particle_id*df.fileid

    max_frame = np.max(df.frame)+1
    df['unique_frame_id'] = df.frame + max_frame*df.unique_particle_id

    clean_df = df[df.channel==0][['unique_particle_id', 'unique_frame_id', 'frame', 'x', 'y', 'z']]
    clean_df.rename({'unique_particle_id' : 'id'}, axis='columns', inplace=True)
    clean_df.reset_index(drop=True, inplace=True)

    assert np.all(df[df.channel==1].unique_frame_id.to_numpy() == clean_df.unique_frame_id)
    for dim in ['x', 'y', 'z']:
        clean_df[dim] -= df[df.channel==1][dim].to_numpy()
        clean_df[dim] *= pixelsize[dim]

    return clean_df

def load_TS(filename, pixelsize, tags):
    df = load_clean_df(filename, pixelsize)
    with io.StringIO(df.to_csv()) as csv_stream:
        return tl.io.load.csv(csv_stream,
                              [None, 'id', None, 't', 'x', 'y', 'z'],
                              delimiter=',',
                              skip_header=1,
                              tags=tags,
                             )

labels_and_files = [
    ('C27', 'ML_trajectories_C27_from_4D_model_2021_04_01_v1_doubleClassification.csv'),
    ('CTCF', 'ML_trajectories_CTCF_from_4D_model_2021_04_01_v1_doubleClassification.csv'),
    ('Rad21', 'ML_trajectories_Rad21_from_4D_model_2021_04_01_v2_doubleClassification.csv'),
    ('C65', 'ML_trajectories_C65_from_4D_model_2021_04_01_v1_doubleClassification.csv'),
    ('C36', 'ML_trajectories_C36_from_4D_model_2021_04_01_v1_doubleClassification.csv'),
#     'ML_trajectories_Rad21_from_4D_model_2021_04_01_v1_doubleClassification.csv',
#     'ML_trajectories_from_4D_model_2021_04_01_v2_doubleClassification.csv',
]
pixelsize = {'x' : 85, 'y' : 85, 'z' : 250}
# 1 frame = 1sec

data = tl.TaggedSet()
for label, filename in labels_and_files:
    new_data = load_TS(filename, pixelsize, label)
    print(f"Loaded {len(new_data)} {label} trajectories")
    data |= new_data
del new_data
print( "-------------------------------------")
print(f"Loaded {len(data)} trajectories total")

data.makeSelection()
tagset = data.tagset()
print('Tags:', tagset)
```

### Gaps

```python
def gaps(traj):
    valid_times = np.where(~np.isnan(traj.abs()[:][:, 0]))[0]
    gaps = np.diff(valid_times)
    return gaps#[gaps > 1]

data.makeSelection()
all_gaps = []
for traj in data:
    all_gaps += gaps(traj).tolist()
    
bins = np.arange(110)+0.5
binmids = (bins[1:] + bins[:-1])/2
plt.hist(all_gaps, bins=bins)
plt.plot(binmids, 2e5*binmids**(-2.6), color='r', label='2e5*L^(-2.6)')
plt.yscale('log')
plt.xscale('log')
plt.xlim([min(bins), max(bins)])
plt.ylim([1, 1e6])
plt.legend()
plt.title('Gap lengths')
plt.xlabel('length L of gaps in frames')
plt.ylabel('count')
plt.show()
```

```python
def split_trajectory_at_long_gaps(traj, max_gapsize):
    valid_times = np.where(~np.isnan(traj.abs()[:][:, 0]))[0]
    gaps = np.diff(valid_times)
    long_gaps = np.where(gaps > max_gapsize)[0]
    
    new_trajs = set()
    t_cur = 0
    for ind_gap in long_gaps:
        new_trajs.add(tl.Trajectory.fromArray(traj.data[:, t_cur:(valid_times[ind_gap]+1), :]))
        t_cur = valid_times[ind_gap+1]
    new_trajs.add(tl.Trajectory.fromArray(traj.data[:, t_cur:, :]))
    
    return new_trajs

def split_ds_at_long_gaps(data, max_gapsize):
    def gen():
        for traj, tags in data(giveTags=True):
            for part_traj in split_trajectory_at_long_gaps(traj, max_gapsize):
                yield(deepcopy(part_traj), deepcopy(tags))
    return tl.TaggedSet(gen())
```

```python
# Choose settings
min_len = 50
min_p_okay = 0.5

def len_okay_selector(traj, _):
    long_enough = len(traj) >= min_len
    enough_data = traj.valid_frames() / len(traj) >= min_p_okay
    return long_enough and enough_data

data.makeSelection(selector=len_okay_selector)
print(f"min_len={min_len}\nmin_p_okay={min_p_okay}\n")
print(f"Still have {len(data)} trajectories")
for tag in tagset:
    data.makeSelection(selector=len_okay_selector)
    data.refineSelection(tags=tag)
    print("{:6d} {:6s} ({:5d} valid frames)".format(len(data), tag, np.sum([traj.valid_frames() for traj in data])))
```

```python
# Cut!
data.makeSelection(selector=lambda traj, tags : not len_okay_selector(traj, tags))
data.deleteSelection()
print(len(data))
```

```python
def twoLocusMSD(t, Gamma, tau_c):
    """
    t : time coordinate
    Gamma : polymer prefactor
    tau_c : crossover-time to constant
    
    Note: we write 2Γ as prefactor, since Γ should be the anomalous
          diffusion coefficient of a single point on the polymer.
    """
    with np.errstate(divide='ignore'):
        ret = 2 * Gamma * (np.sqrt(t)*(1-np.exp(-tau_c/(np.pi*t))) + np.sqrt(tau_c)*scipy.special.erfc(np.sqrt(tau_c/(np.pi*t))))
    ret[np.isnan(ret)] = 0
    return ret

pixelsize = {'x' : 85, 'y' : 85, 'z' : 250}
def traj_noise_from_aggregate(noise2):
    voxel = np.array([pixelsize[dim] for dim in ['x', 'y', 'z']])
    noise_in_px = np.sqrt(noise2/np.sum(voxel**2))
    return noise_in_px*voxel
```

```python
# For localization error estimation we use only C36
msdfits = {}
data.makeSelection()
for tag in data.tagset():
    data.makeSelection(tags=tag)
    
    msd, N = tl.analysis.MSD(data, giveN=True)
    msd *= 1e-6
    t_msd = 20*np.arange(len(msd))

    popt, pcov = scipy.optimize.curve_fit(lambda x, G, tau, noise2 : np.log(twoLocusMSD(x, G, tau) + 2*noise2),
                                          t_msd[1:], np.log(msd[1:]), p0=(4e-3, 8e2, 0.04),
                                          sigma = 1/np.sqrt(N[1:]), bounds=(0, np.inf))

    if tag == 'CTCF': # Fit screws up for some reason, so use handpicked parameters
        popt = np.array([4e-3, 8e2, 0.04])
        pcov = np.array([[np.inf]])
        
    msdfits[tag] = popt
    traj_noise = traj_noise_from_aggregate(popt[2])

    plt.loglog(t_msd[1:], msd[1:], label=tag)
    plt.loglog(t_msd[1:], twoLocusMSD(t_msd[1:], *popt[:2]) + 2*popt[2], color='r', label='fit')
    plt.loglog(t_msd[1:], twoLocusMSD(t_msd[1:], *popt[:2]), color='m', label='Rouse')
    plt.axhline(2*popt[2], color='c', label='error')

    plt.legend()
    plt.title('Fitting Rouse + localization error')
    plt.xlabel('time [seconds]')
    plt.ylabel('MSD [μm^2]')
    plt.show()

    print("Fit results")
    print("-----------")
    print("  Γ = {:.3g} μm^2/√s".format(popt[0]))
    print("  τ = {:.3g} s ==> rms(R_e2e) = {:.3g} μm".format(popt[1], np.sqrt(popt[0]*np.sqrt(popt[1]))))
    print("  noise = {:.3g} μm^2".format(popt[2]))
    print("    --> error on distance vectors: [{:.3f}, {:.3f}, {:.3f}]μm".format(*traj_noise))
    print("    --> error on single spots:     [{:.3f}, {:.3f}, {:.3f}]μm".format(*traj_noise/np.sqrt(2)))
    print("")
    print("relative uncertainty: {:.3g}".format(( np.sqrt(np.linalg.det(pcov)) / np.prod(popt) )**(1/3)))
    print("")
```

```python
# Write localization errors into trajectories
for tag, popt in msdfits.items():
    traj_noise = 1e3*traj_noise_from_aggregate(popt[2]) # trajectories are in nm
    data.makeSelection(tags=tag)
    for traj in data:
        traj.meta['localization_error'] = deepcopy(traj_noise)
```

### Calibration

```python
# modelfam = neda.ParametricFamily((0.2, 10, 3), [(1e-10, None), (1e-10, None), (1e-10, None)])
# modelfam.get = lambda D, k, k_extra : neda.models.RouseModel(20, D, k, k_extra=k_extra, looppositions=[(0, 0), (0, -1)])

modelfam = neda.ParametricFamily((170, 350), [(1e-10, None), (1e-10, None)])
modelfam.get = lambda s1, s0 : neda.models.FactorizedModel([scipy.stats.maxwell(scale=s0),
                                                            scipy.stats.maxwell(scale=s1),
                                                           ])

data.makeSelection(tags=['Rad21', 'C27'], logic=any)
fitres = neda.models.fit(data, modelfam)
```

```python
np.set_printoptions(precision=3)
# print("Fitted Rouse parameters")
# print("D = {:.3g}, k = {:.3g}, k_extra = {:.3g}".format(*fitres.x))
print("Fitted factorized parameters")
print("s1 = {:.3g}, s0 = {:.3g}".format(*fitres.x))
print()
print("relative uncertainty: {:.3g}".format((np.sqrt(np.linalg.det(fitres.hess_inv.todense()))/np.prod(fitres.x))**(1/3)))
stds = np.sqrt(np.diagonal(fitres.hess_inv.todense()))
print("individual stds: ", stds)
print("correlation matrix:")
print(fitres.hess_inv.todense() / (stds[:, np.newaxis]*stds[np.newaxis, :]))
np.set_printoptions()
```

```python
rouseparams = fitres.x
# model = neda.models.RouseModel(20, rouseparams[0], rouseparams[1], k_extra=rouseparams[2], looppositions=[(0, 0), (0, -1)])

model = modelfam.get(mwparam['C27'][1], mwparam['Rad21'][1])

priorfam = neda.ParametricFamily((0,), [(None, 0)])
priorfam.get = lambda logq : neda.priors.GeometricPrior(logq, nStates=2)
MCMCconfig = {
    'iterations' : 3000,
    'burn_in' : 500,
}

def parfun(params):
    seed, traj = params
    np.random.seed(seed)
    return neda.main(traj, model, priorfam, MCMCconfig, return_='traj')

data.makeSelection(tags='C36')
with Pool(18) as mypool:
    zipped = zip(np.arange(len(data))+132, data)
    imap = mypool.imap_unordered(parfun, zipped)
    data_C36inf_fac = tl.TaggedSet(tqdm(imap, total=len(data)), hasTags=False)
```

### Loop lifetime distribution

```python
def looplifes_lt(lt, state):
        states_full = lt.full_valid()
        states_padded = np.pad(states_full, (1, 1), constant_values=-1)
        starts = np.where((states_padded[:-1] != state) & (states_padded[1:] == state))[0]
        ends   = np.where((states_padded[:-1] == state) & (states_padded[1:] != state))[0]
        return ends-starts

def looplifes(traj, step=1, state=1):
    lifetimes = []
    for lt in traj.meta['neda']['final']['mcmcrun'].samples[::step]:
        lifetimes += looplifes_lt(lt, state).tolist()
        
    return np.array(lifetimes)
```

```python
data_inf = data_C36inf_fac
looplifetimes = {'looped' : [], 'unlooped' : []}
data_inf.makeSelection()

def parfun0(traj):
    return looplifes(traj, state=0, step=5)
def parfun1(traj):
    return looplifes(traj, state=1, step=5)
with Pool(18) as mypool:
    for life in tqdm(mypool.imap_unordered(parfun0, data_inf), total=len(data)):
        looplifetimes['unlooped'] += life.tolist()
    for life in tqdm(mypool.imap_unordered(parfun1, data_inf), total=len(data)):
        looplifetimes['looped'] += life.tolist()
```

```python
for status in ['looped', 'unlooped']:
    plt.hist(looplifetimes[status], bins=np.arange(100)+0.5, density=True, alpha=0.7, label=status)
    
plt.yscale('log')
plt.legend()
plt.title('Inferred interval lengths (C36, factorized likelihood, manual calibration)')
plt.xlabel('interval length in frames')
plt.ylabel('density')
plt.ylim([1e-4, 2e-1])
plt.show()

def sem(x):
    return np.std(x) / np.sqrt(len(x))

print('mean looped lifetime:   {:.3g} +- {:.3g} frames'.format(np.mean(looplifetimes['looped'  ]), sem(looplifetimes['looped'  ])))
print('mean unlooped lifetime: {:.3g} +- {:.3g} frames'.format(np.mean(looplifetimes['unlooped']), sem(looplifetimes['unlooped'])))
```

## KLD

```python
# Count good snippets in each dataset
def n_good_snippets(traj, n=10):
    t_valid = np.where(~np.isnan(traj.abs()[:][:, 0]))[0]
    # Count how often we have a window of n data points spanning n frames (i.e. no missing frames)
    return np.count_nonzero(np.diff(t_valid, n=n) == n)

for n in [2, 5, 10, 20]:
    data.makeSelection()
    print("For snippets of size {} we have".format(n))
    for tag in data.tagset():
        data.makeSelection(tags=tag)
        n_snip = np.sum([n_good_snippets(traj, n) for traj in data])
        print("{:5s} dataset: {:5d} good snippets".format(tag, n_snip))
    print()
```
